```python
import os
import time
import warnings
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from IPython.display import display, Image, HTML

from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.metrics import (roc_auc_score, accuracy_score, f1_score,
                            confusion_matrix, roc_curve, auc, classification_report,
                            precision_score, recall_score)
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

warnings.filterwarnings("ignore")
# checking the csv file
def mkdir_p(p):
    os.makedirs(p, exist_ok=True)

def try_read_csv(path):
    seps = [',', ';', '\t']
    for sep in seps:
        try:
            df = pd.read_csv(path, sep=sep, engine="python", on_bad_lines="skip")
            if df.shape[1] > 1:
                return df
        except Exception:
            continue
    try:
        return pd.read_csv(path, engine="python", on_bad_lines="skip")
    except Exception:
        print(f"Error reading file: {path}")
        return pd.DataFrame()
# encoding and clean
def OHE_friendly():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)

def freq_encode_one(s: pd.Series):
    return s.value_counts(dropna=False, normalize=True)

def basic_clean(df, target_col):
    df.columns = [c.strip().replace(' ', '_').lower() for c in df.columns]

    drop_these = [c for c in df.columns if any(x in c for x in ("id", "name", "town", "customer"))]
    if drop_these:
        df = df.drop(columns=drop_these, errors="ignore")

    for c in df.select_dtypes(include=['object']).columns:
        df[c] = df[c].astype(str).str.strip().str.lower()
        df[c] = df[c].replace({'unknown': np.nan, 'none': np.nan, 'na': np.nan})

    if target_col in df.columns:
        df[target_col] = df[target_col].astype(str).str.lower().map({'yes': 1, 'no': 0})
        df = df[df[target_col].isin([0, 1])].copy()

    return df
# plots and bossters
def small_eda_save(df, outdir, target_col):
    mkdir_p(outdir)
    sns.set(style="whitegrid")

    path_counts = os.path.join(outdir, "target_counts.png")
    fig, ax = plt.subplots(figsize=(6,3))
    sns.countplot(x=target_col, data=df, ax=ax)
    ax.set_title("Target Distribution")
    fig.savefig(path_counts)
    plt.close(fig)

    path_missing = os.path.join(outdir, "missing_heatmap.png")
    sample = df if len(df) <= 1500 else df.sample(1500, random_state=1)
    fig, ax = plt.subplots(figsize=(10,4))
    sns.heatmap(sample.isnull(), cbar=False, yticklabels=False, ax=ax)
    ax.set_title("Missing Data Heatmap")
    fig.savefig(path_missing)
    plt.close(fig)
    return path_counts, path_missing

def save_plot(fig, path):
    fig.tight_layout()
    fig.savefig(path, bbox_inches='tight')
    plt.close(fig)

def build_preproc(Xtrain, cardinal_threshold=20):
    num_cols = Xtrain.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in Xtrain.columns if c not in num_cols]

    low_card = [c for c in cat_cols if Xtrain[c].nunique(dropna=False) <= cardinal_threshold]

    num_pipe = Pipeline([
        ('imp_med', SimpleImputer(strategy='median')),
        ('scale', StandardScaler())
    ])

    cat_pipe = Pipeline([
        ('imp_cat', SimpleImputer(strategy='constant', fill_value='missing')),
        ('ohe', OHE_friendly())
    ])

    preproc = ColumnTransformer([
        ('num', num_pipe, num_cols),
        ('lowcat', cat_pipe, low_card)
    ], remainder='passthrough')

    high_card = [c for c in cat_cols if c not in low_card]
    return preproc, num_cols, low_card, high_card

def get_feature_names(preproc, Xtrain, low_card, high_card):

    try:
        # More robust feature name extraction
        num_features = preproc.named_transformers_['num'].get_feature_names_out()
        ohe_features = preproc.named_transformers_['lowcat']['ohe'].get_feature_names_out(low_card)

        # Combine all feature names
        feature_names = list(num_features) + list(ohe_features) + high_card
        return feature_names
    except Exception as e:
        print(f"Warning: Could not extract feature names properly: {e}")
        # Fallback to basic feature names
        num_cols = Xtrain.select_dtypes(include=[np.number]).columns.tolist()
        return num_cols + low_card + high_card

def mk_pipe(clf, preproc, use_smote):
    steps = [('pre', preproc)]
    if use_smote:
        steps.append(('sm', SMOTE(random_state=42)))
    steps.append(('clf', clf))
    return ImbPipeline(steps)

def safe_predict(pipe, Xraw):
    y_pred, y_proba = pipe.predict(Xraw), None
    try:
        y_proba = pipe.predict_proba(Xraw)[:, 1]
    except Exception:
        try:
            df = pipe.decision_function(Xraw)
            y_proba = (df - df.min()) / (df.max() - df.min() + 1e-9)
        except Exception:
            pass
    return y_pred, y_proba

def compute_metrics(y_true, y_pred, y_proba):
    d = {
        'accuracy': accuracy_score(y_true, y_pred),
        'f1': f1_score(y_true, y_pred, zero_division=0),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall': recall_score(y_true, y_pred, zero_division=0)
    }
    try:
        if y_proba is not None:
            d['roc_auc'] = roc_auc_score(y_true, y_proba)
    except Exception:
        d['roc_auc'] = np.nan

    return {k: float(v) for k, v in d.items()}
# plots
def main(args):
    start = time.time()
    random.seed(args.seed)
    np.random.seed(args.seed)

    mkdir_p(args.output_dir)
    plots_dir = os.path.join(args.output_dir, "plots")
    mkdir_p(plots_dir)
    fimp_dir = os.path.join(args.output_dir, "feature_maps")
    mkdir_p(fimp_dir)

    print(f"Reading input: {args.input}")
    df = try_read_csv(args.input)
    if df.empty:
        print("Input data could not be read. Exiting.")
        return

    df = basic_clean(df, args.target)
    print(f"Data rows after cleaning: {len(df)}")

    path_counts, path_missing = small_eda_save(df, plots_dir, args.target)

    print("\nEDA Plots:")
    display(Image(path_counts))
    display(Image(path_missing))

    X = df.drop(columns=[args.target]).copy()
    y = df[args.target].astype(int)
    Xtrain, Xtest, ytrain, ytest = train_test_split(
        X, y,
        test_size=args.test_size,
        random_state=args.seed,
        stratify=y
    )
    print(f"\nTrain/Test split: {len(Xtrain)} / {len(Xtest)}")

    preproc, num_cols, low_card, highcats = build_preproc(Xtrain, cardinal_threshold=args.cardinal_threshold)

    Xtrain_fe, Xtest_fe = Xtrain.copy(), Xtest.copy()
    freq_maps = {}
    for c in highcats:
        freqs = freq_encode_one(Xtrain_fe[c])
        freq_maps[c] = freqs
        Xtrain_fe[c] = Xtrain_fe[c].map(freqs).fillna(0.0)
        Xtest_fe[c] = Xtest_fe[c].map(freqs).fillna(0.0)
    print(f"Frequency encoded {len(highcats)} high-cardinality columns.")

    preproc.fit(Xtrain_fe)

    # Get final feature names after preprocessing and frequency encoding
    feature_names = get_feature_names(preproc, Xtrain, low_card, highcats)

    models = {}
    rs_cv = StratifiedKFold(n_splits=args.cv_folds, shuffle=True, random_state=42)

    print("\nTraining Models:")

    # 1. XGBoost
    xgb_clf = XGBClassifier(
        tree_method='hist', random_state=42, n_jobs=-1, n_estimators=150, max_depth=6, learning_rate=0.05
    )
    pipe_xgb = mk_pipe(xgb_clf, preproc, args.use_smote)
    pipe_xgb.fit(Xtrain_fe, ytrain)
    models["XGBoost"] = pipe_xgb
    print("XGBoost fitted.")

    # 2. LightGBM
    lgb_clf = LGBMClassifier(
        n_estimators=150, num_leaves=31, learning_rate=0.05, random_state=42, n_jobs=-1, verbose=-1  # FIXED: Added verbose=-1 to suppress warnings
    )
    pipe_lgb = mk_pipe(lgb_clf, preproc, args.use_smote)
    pipe_lgb.fit(Xtrain_fe, ytrain)
    models["LightGBM"] = pipe_lgb
    print("LightGBM fitted.")

    # 3. Tuned Random Forest
    rf_pipe = mk_pipe(RandomForestClassifier(random_state=42, n_jobs=-1), preproc, args.use_smote)
    rf_grid = {
        "clf__n_estimators": [100, 200], "clf__max_depth": [None, 10], "clf__min_samples_split": [2, 5]
    }
    rs = RandomizedSearchCV(
        rf_pipe, rf_grid, n_iter=max(1, args.rs_iter), scoring='roc_auc', cv=rs_cv, n_jobs=-1, random_state=42
    )
    rs.fit(Xtrain_fe, ytrain)
    models["RandomForest_Tuned"] = rs.best_estimator_
    print("Random Forest tuned and fitted.")

    # 4. Tuned MLP
    mlp_pipe = mk_pipe(MLPClassifier(max_iter=400, random_state=42), preproc, args.use_smote)
    mlp_grid = {
        "clf__hidden_layer_sizes": [(64,), (128, 64)], "clf__alpha": [1e-4, 1e-3], "clf__learning_rate_init": [1e-3]
    }
    ms = RandomizedSearchCV(
        mlp_pipe, mlp_grid, n_iter=max(1, args.rs_iter), scoring='roc_auc', cv=rs_cv, n_jobs=-1, random_state=42
    )
    ms.fit(Xtrain_fe, ytrain)
    models["MLP_Tuned"] = ms.best_estimator_
    print("MLP (Sklearn) tuned and fitted.")

    print("\nModel Evaluation:")
    results = {}
    plot_paths = []

    for name, pipe in models.items():
        y_pred, y_proba = safe_predict(pipe, Xtest_fe)

        if y_pred is None:
            continue

        metrics = compute_metrics(ytest, y_pred, y_proba)
        results[name] = metrics
        print(f"  {name}: ROC AUC = {metrics.get('roc_auc', 'N/A'):.4f}")

        # Display Classification Report (a map of precision/recall/f1)
        print(f"\n--- Classification Report: {name} ---")
        report = classification_report(ytest, y_pred, output_dict=True, zero_division=0)
        report_df = pd.DataFrame(report).transpose().round(4)
        print(report_df)

        # Feature Importance Map (for tree-based models only)
        if hasattr(pipe.named_steps['clf'], 'feature_importances_'):
            importances = pipe.named_steps['clf'].feature_importances_

            # Handle length mismatch between features and importances
            if len(feature_names) == len(importances):
                fimp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
            else:
                print(f"Warning: Feature name mismatch ({len(feature_names)} names vs {len(importances)} importances)")
                fimp_df = pd.DataFrame({
                    'Feature': [f'feature_{i}' for i in range(len(importances))],
                    'Importance': importances
                })

            fimp_df = fimp_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)

            fimp_path_csv = os.path.join(fimp_dir, f"{name}_feature_importance.csv")
            fimp_df.to_csv(fimp_path_csv, index=False)
            print(f"Feature Importance saved to: {fimp_path_csv}")

            # Plot Feature Importance
            fig, ax = plt.subplots(figsize=(8, 6))
            sns.barplot(x='Importance', y='Feature', data=fimp_df.head(10), ax=ax)
            ax.set_title(f"Top 10 Feature Importance: {name}")
            fimp_path_png = os.path.join(plots_dir, f"{name}_fimp.png")
            save_plot(fig, fimp_path_png)
            plot_paths.append(fimp_path_png)

        # Confusion Matrix Plot
        cm = confusion_matrix(ytest, y_pred)
        fig, ax = plt.subplots(figsize=(4, 3))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
        ax.set_title(f"Confusion: {name}")
        path = os.path.join(plots_dir, f"conf_{name}.png")
        save_plot(fig, path)
        plot_paths.append(path)

        # ROC Curve Plot
        if 'roc_auc' in metrics and not np.isnan(metrics['roc_auc']):
            fpr, tpr, _ = roc_curve(ytest, y_proba)
            fig, ax = plt.subplots(figsize=(6, 4))
            ax.plot(fpr, tpr, label=f"AUC={auc(fpr, tpr):.3f}")
            ax.plot([0, 1], [0, 1], linestyle='--', color='grey')
            ax.set_title(f"ROC Curve: {name}")
            ax.legend()
            path = os.path.join(plots_dir, f"roc_{name}.png")
            save_plot(fig, path)
            plot_paths.append(path)

    metrics_df = pd.DataFrame(results).T
    metrics_df.to_csv(os.path.join(args.output_dir, "metrics_final.csv"))

    fig, ax = plt.subplots(figsize=(8, max(3, 0.5 * len(metrics_df))))
    vals = metrics_df['roc_auc'].fillna(0).sort_values(ascending=False)
    sns.barplot(x=vals.values, y=vals.index, ax=ax)
    ax.set_xlabel('ROC AUC')
    ax.set_title("Model Comparison by ROC AUC")
    cmp_path = os.path.join(plots_dir, "cmp_roc_auc.png")
    save_plot(fig, cmp_path)
    plot_paths.append(cmp_path)

    print("\nFinal Metrics Table:")
    display(metrics_df.nlargest(len(models), 'roc_auc').style.format({
        'accuracy': '{:.4f}',
        'f1': '{:.4f}',
        'roc_auc': '{:.4f}',
        'precision': '{:.4f}',
        'recall': '{:.4f}'
    }))

    print("\nComparison Plot:")
    display(Image(cmp_path))

    print("\nAll Generated Maps (Plots):")
    for p in plot_paths:
        display(Image(p))

    print("\nSaving models and artifacts...")
    joblib.dump(preproc, os.path.join(args.output_dir, "preprocessor.joblib"))
    joblib.dump(freq_maps, os.path.join(args.output_dir, "freq_maps.joblib"))

    for nm, pipe in models.items():
        joblib.dump(pipe, os.path.join(args.output_dir, f"{nm}_pipe.joblib"))

    elapsed = time.time() - start
    print(f"\nDone! Outputs saved to: {args.output_dir}")
    print(f"Elapsed time: {int(elapsed)} seconds")


# Configuration class for easy parameter adjustment
class NotebookArgs:
    input = "wallacecommunications.csv"
    output_dir = "output_colab_full_maps"
    test_size = 0.2
    seed = 42
    rs_iter = 2
    use_smote = False
    cardinal_threshold = 20
    target = "new_contract_this_campaign"
    cv_folds = 3

# Run the pipeline
if __name__ == "__main__":
    args = NotebookArgs()
    main(args)
```
